```python
def plot_rmsf(folder, replica):    
    """Function to read the rmsf record from the MD simulation for all replica.
    
    Parameters
    ----------
    folder: str
        path to data.
    replica: int
        number of replicas.
        
    Returns
    ------
    rmsf_df: pandas DataFrame
        With the RMSF record in nm.
    """
    
    rmsf=[]
    for i in np.arange(1,replica+1,1):
        rmsf_protein_n, rmsf_dna_n = plotting_gmx.plot_ana_rmsf(folder+'{}/analysis/ana/'.format(i))
        rmsf_protein_n['Replica'] = [i for x in range(len(rmsf_protein_n))]
        rmsf_protein_n['Molecule'] = ["Protein" for x in range(len(rmsf_protein_n))]
        rmsf_dna_n['Replica'] = [i for x in range(len(rmsf_dna_n))]
        rmsf_dna_n['Molecule'] = ["DNA" for x in range(len(rmsf_dna_n))]
        rmsf.append(rmsf_protein_n)
        rmsf.append(rmsf_dna_n)
    plt.show()
    rmsf_df = pd.concat([df for df in rmsf])
    
    return(rmsf_df)

rmsf_df = plot_rmsf(folder_OSymL,3)
rmsf_df_NOD = plot_rmsf(folder_NOD,3)
```

```python
sns.lineplot(data=rmsf_df[rmsf_df["Molecule"]=="Protein"], x="Residue", y = "RMSF (nm)", color="purple")
sns.lineplot(data=rmsf_df_NOD[rmsf_df_NOD["Molecule"]=="Protein"], x="Residue", y = "RMSF (nm)", color="orange")

plt.show()
sns.lineplot(data=rmsf_df[rmsf_df["Molecule"]=="DNA"], x="Residue", y = "RMSF (nm)", color="purple")
sns.lineplot(data=rmsf_df_NOD[rmsf_df_NOD["Molecule"]=="DNA"], x="Residue", y = "RMSF (nm)", color="orange")

plt.show()
```

```python
def get_Etot(path, folder=folder_OSymL, rep=False):    
    """Function that read the total energy output from GROMACS into a DataFrame.
    
    Total energy can be calculated with gmx_mpi energy -f ../md_NMR.edr -o Etot.xvg
    
    """
    
    Etot = pd.read_csv(folder + "1/{}Etot.xvg".format(path), sep='\s+',skiprows=24,names=['Frame','E$_{total}$ (kJ/mol)'])
    Etot['replica'] = [1 for x in range(len(Etot))]
    Etot['Time (ns)'] = Etot["Frame"]/1000

    if rep==True:
        for i in [2,3]:
            Etot_n = pd.read_csv(folder + "{0}/{1}Etot.xvg".format(i,path), \
                                   sep='\s+',skiprows=24,names=['Frame','E$_{total}$ (kJ/mol)'])

            Etot_n['replica'] = [i for x in range(len(Etot_n))]
            Etot_n['Time (ns)'] = Etot_n["Frame"]/1000

            Etot = Etot.append(Etot_n)

    return(Etot)
```

```python
Ener = get_Etot('analysis/ana/', rep=True)
Ener_NOD = get_Etot('analysis/ana/', folder=folder_NOD, rep=True)
plt.subplots(figsize=(2.5,1.8))

m, s =Ener.groupby("Time (ns)").mean().rolling(1000).mean()[::100],\
Ener.groupby("Time (ns)").std().rolling(1000).mean()[::100]

m_NOD, s_NOD =Ener_NOD.groupby("Time (ns)").mean().rolling(1000).mean()[::100],\
Ener_NOD.groupby("Time (ns)").std().rolling(1000).mean()[::100]

plt.plot(m.index, m["E$_{total}$ (kJ/mol)"], color="purple")

plt.fill_between(m.index, m["E$_{total}$ (kJ/mol)"]-s["E$_{total}$ (kJ/mol)"],\
                 y2=m["E$_{total}$ (kJ/mol)"]+s["E$_{total}$ (kJ/mol)"], alpha = 0.5, color="purple")

plt.plot(m_NOD.index, m_NOD["E$_{total}$ (kJ/mol)"], color="orange")

plt.fill_between(m_NOD.index, m_NOD["E$_{total}$ (kJ/mol)"]-s_NOD["E$_{total}$ (kJ/mol)"],\
                 y2=m_NOD["E$_{total}$ (kJ/mol)"]+s_NOD["E$_{total}$ (kJ/mol)"], alpha = 0.5, color="orange")

plt.ylabel("E$_{total}$ (kJ/mol)")
plt.xlabel("Time (ns)")
plt.tight_layout()

plt.savefig(folder+"/Energy.pdf", dpi=300)
plt.show()
```

```python
def get_viol(path, folder=folder_OSymL, rep=False):    
    avg_viol = pd.read_csv(folder + "1/{}draver.xvg".format(path), sep='\s+',skiprows=18,names=['Frame','Viol'])
    avg_viol['replica'] = [1 for x in range(len(avg_viol))]
    avg_viol['Time (ns)'] = avg_viol["Frame"]/1000

    max_viol = pd.read_csv(folder + "1/analysis/ana/drmax.xvg", sep='\s+',skiprows=18,names=['Frame','Viol'])
    max_viol['Time (ns)'] = avg_viol["Frame"]/1000
    max_viol['replica'] = [1 for x in range(len(avg_viol))]
    if rep==True:
        for i in [2,3]:
            avg_viol_n = pd.read_csv(folder + "{0}/{1}draver.xvg".format(i, path), \
                                   sep='\s+',skiprows=18,names=['Frame','Viol'])

            avg_viol_n['replica'] = [i for x in range(len(avg_viol_n))]
            avg_viol_n['Time (ns)'] = avg_viol_n["Frame"]/1000

            avg_viol = avg_viol.append(avg_viol_n)

            max_viol_n = pd.read_csv(folder + "{0}/{1}drmax.xvg".format(i, path), \
                                   sep='\s+',skiprows=18,names=['Frame','Viol'])

            max_viol_n['replica'] = [i for x in range(len(max_viol_n))]
            max_viol_n['Time (ns)'] = avg_viol_n["Frame"]/1000
            max_viol = avg_viol.append(max_viol_n)
    return(avg_viol, max_viol)
```
